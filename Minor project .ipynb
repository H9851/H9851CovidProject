```python
import pandas as pd
import numpy as np
import seaborn as sn
from matplotlib import pyplot as pl
from matplotlib import style
from datetime import date
```

```python
#Read CSV File
data_csv = pd.read_csv("covid_19_india.csv")
```

```python
data_csv.info()
```

```python
pd.read_csv('covid_19_india.csv', nrows = 3711)
```

```python
dframe_state_data = pd.pivot_table(data_csv, values = ['Confirmed', 'Cured', 'Deaths'], index = 'State/UnionTerritory', aggfunc = 'max')
dframe_state_data['Death Rate'] = round(dframe_state_data['Deaths'] / dframe_state_data['Confirmed'], 2)
dframe_state_data = dframe_state_data.sort_values(by = 'Confirmed', ascending = False).reset_index(level = 0)
dframe_state_data.style.background_gradient(cmap = 'Spectral')
```

# Covid data set of Delhi


```python
#Read CSV File
data_csv_delhi = data_csv[data_csv['State/UnionTerritory'] == "Delhi"]
```

```python
data_csv_delhi.info()
```

```python
data_csv_delhi.describe()
```

```python
data_csv_delhi = pd.read_csv('covid_19_india.csv', index_col = 1, parse_dates = True, dayfirst = True)
start_date = '2020-06-01'
end_date   = '2020-06-30'
data_csv_delhi = data_csv_delhi.loc[start_date : end_date]
data_csv_delhi2 = data_csv_delhi.loc[(data_csv_delhi['State/UnionTerritory'] == 'Delhi')]
data_csv_delhi = pd.DataFrame(data_csv_delhi2)
data_csv_delhi = data_csv_delhi.reset_index()
data_csv_delhi.get(['Sno', 'Date', 'Time', 'State/UnionTerritory', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed'])
for data in data_csv_delhi.index:
    data_csv_delhi['Sno'][data], data_csv_delhi['Date'][data], data_csv_delhi['Time'][data], data_csv_delhi['State/UnionTerritory'][data], data_csv_delhi['ConfirmedIndianNational'][data], data_csv_delhi['ConfirmedForeignNational'][data], data_csv_delhi['Cured'][data], data_csv_delhi['Deaths'][data], data_csv_delhi['Confirmed'][data]
data_csv_delhi
```

# Using matplotbit to Visualise  Covid data of delhi

```python
dateList = [data.strftime("%d") for data in data_csv_delhi['Date'].tolist()]
pl.plot(dateList, data_csv_delhi['Confirmed'].tolist(), color = 'green', label = 'Confirmed', linewidth = 5)
pl.plot(dateList, data_csv_delhi['Cured'].tolist(), color = 'blue', label = 'Cured', linewidth = 5)
pl.plot(dateList, data_csv_delhi['Deaths'].tolist(), color = 'red', label = 'Deaths', linewidth = 5)
pl.title('COVID - 19 Delhi')
pl.xlabel('Days of June')
pl.ylabel('Number Of People')
pl.legend()
pl.grid(True, color = 'gray')
pl.legend(loc = 'best')
pl.show()
```

# Bar Chart Delhi Cases

```python
fig, ax = pl.subplots()
data_csv_delhi.plot(kind = 'bar', x = 'Date', y = 'Confirmed', color = 'green', figsize = (20, 10), ax = ax)
data_csv_delhi.plot(kind = 'bar', x = 'Date', y = 'Cured', color = 'blue', figsize = (20, 10), ax = ax)
data_csv_delhi.plot(kind = 'bar', x = 'Date', y = 'Deaths', color = 'red', figsize = (20, 10), ax = ax)
ax.set_title('COVID - 19 Delhi')
ax.set_xlabel('Days Of June')
ax.set_ylabel('Number Of People')
pl.show()
```

# Pie CHart to visualise the Data of Delhi 

```python
fig, ax = pl.subplots()
group1 = [85161, 56235, 2680]
group2 = [0.1, 0.1, 0]
group3 = ['Confirmed', 'Cured', 'Deaths']
ax.pie(group1, explode = group2, labels = group3, autopct = '%1.2f%%', shadow = True, startangle = 90)
ax.axis('Equal')
ax.set_title('COVID - 19 Delhi')
pl.show()
```
